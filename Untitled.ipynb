```python
# open Kong Oscar 2020 dataset
file_dir = Path('C:/Users/marie/Documents/PhD/Chapter_3/data/OMG_Narwhals_mooring_data/OMG_Narwhals_mooring_L2/2020')
kongoscar20_ds = xr.open_dataset(file_dir / 'OMG_Narwhals_Mooring_CTD_L2_03716787_20190804.nc') # shallow
# kongoscar20_ds = xr.open_dataset(file_dir / 'OMG_Narwhals_Mooring_CTD_L2_03716790_20190804.nc') # deep
kongoscar20_ds.close()
```

```python
kongoscar20_temp_avg = kongoscar20_ds.temperature.resample(time='1D').mean(dim=["time"]) # average
kongoscar20_temp_std = kongoscar20_ds.temperature.resample(time='1D').std(dim=["time"]) # standard dev
```

```python
kongoscar19_temp_avg = kongoscar19_ds.temperature.resample(time='1D').mean(dim=["time"]) # average
kongoscar19_temp_std = kongoscar19_ds.temperature.resample(time='1D').std(dim=["time"]) # standard dev
```

```python
plt.figure(figsize=[11, 5])
plt.rcParams['font.size'] = '12'
plt.margins(x=0)

plt.plot(kongoscar19_temp_avg.time,kongoscar19_temp_avg.values, color='k')
plt.fill_between(kongoscar19_temp_avg.time, kongoscar19_temp_avg.values-kongoscar19_temp_std.values, \
                 kongoscar19_temp_avg.values+kongoscar19_temp_std.values, facecolor='gray', interpolate=True)

plt.plot(kongoscar20_temp_avg.time,kongoscar20_temp_avg.values, color='k')
plt.fill_between(kongoscar20_temp_avg.time, kongoscar20_temp_avg.values-kongoscar20_temp_std.values, \
                 kongoscar20_temp_avg.values+kongoscar20_temp_std.values, facecolor='gray', interpolate=True)

# plt.ylabel("$Q_{net}$ (Wm$^{-2}$)")
# plt.xticks(ticks=heat_months_ordered.month, labels=['M','J','J','A','S','O','N','D','J','F','M','A'])
plt.grid(linestyle='--')
# plt.title("Variation in area-weighted mean monthly net air-sea heat fluxes (May-April)");
```

```python
plt.figure(figsize=[11, 5])
plt.rcParams['font.size'] = '12'
plt.margins(x=0)

plt.plot(kongoscar19_temp_avg.time,kongoscar19_temp_avg.values, color='k')
plt.fill_between(kongoscar19_temp_avg.time, kongoscar19_temp_avg.values-kongoscar19_temp_std.values, \
                 kongoscar19_temp_avg.values+kongoscar19_temp_std.values, facecolor='gray', interpolate=True)

plt.plot(kongoscar20_temp_avg.time,kongoscar20_temp_avg.values, color='k')
plt.fill_between(kongoscar20_temp_avg.time, kongoscar20_temp_avg.values-kongoscar20_temp_std.values, \
                 kongoscar20_temp_avg.values+kongoscar20_temp_std.values, facecolor='gray', interpolate=True)

# plt.ylabel("$Q_{net}$ (Wm$^{-2}$)")
# plt.xticks(ticks=heat_months_ordered.month, labels=['M','J','J','A','S','O','N','D','J','F','M','A'])
plt.grid(linestyle='--')
# plt.title("Variation in area-weighted mean monthly net air-sea heat fluxes (May-April)");
```
